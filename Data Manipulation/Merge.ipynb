```python
import numpy as np
import pandas as pd 
from pandas import Series, DataFrame
```

```python
dframe1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
dframe1
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})
dframe2
```

```python
# merge method 
pd.merge(dframe1,dframe2)
```

```python
pd.merge(dframe1,dframe2,on='key')
#merging on keys
```

```python
pd.merge(dframe1,dframe2,on='key',how='left')
#left dataset key values and index values are being used ... we wanted to save first dataframe
```

```python
dframe2
```

```python
dframe1
```

```python
pd.merge(dframe1,dframe2,on='key',how='right')
# merged on key on dataframe2
```

```python
# union of both keys
pd.merge(dframe1,dframe2,on='key',how='outer')
```

```python
#many to many merge
dframe3 = DataFrame({'key':['X','X','X','Y','Z','Z'],'data_set_3':range(6)})
dframe4 = DataFrame({'key':['Y','Y','X','X','Z'],'data_set_4':range(5)})
```

```python
dframe3
```

```python
dframe4
```

```python
pd.merge(dframe3,dframe4)
# two dataframes are merged based on both their index and they matched on 'key' column valuesare little different
```

```python
#merge with multiple keys
df_left = DataFrame({'key1':['SF','LA','SF'],'key2':['one','two','one'],'left_data':[10,20,30]})
```

```python
df_left
```

```python
df_right = DataFrame({'key1':['SF','LA','SF','LA'],'key2':['one','two','one','two'],'right_data':[40,50,60,70]})
```

```python
df_right
```

```python
#we are merging on multiply keys merging on key1 and key2 on outer so it does union of those
pd.merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_left,df_right,on='key1')
#left and right df have overlaping 'key names' so pandas can automaticly add suffixes to them
```

```python
#pandas added suffix _x and _y 
#we can specify the name of suffixes
pd.merge(df_left,df_right,on='key1',suffixes=('_lefty','_righty'))
```
