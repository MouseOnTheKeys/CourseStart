```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set(style='whitegrid')
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
#pivoting
flight_dframe = flight_dframe.pivot('month','year','passengers')
```

```python
flight_dframe
```

```python
#perfect for heatmap
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_dframe,annot=True,fmt='d')
```

```python
sns.heatmap(flight_dframe,annot=True,fmt='d',center=flight_dframe.loc['January',1955])
```

```python
fig,(axis1,axis2) = plt.subplots(2,1)

yearly_flights=flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

years_dframe = pd.concat((years,flights),axis=1)
years_dframe.columns = ['Years','Flights']

sns.barplot('Years',y='Flights',data=years_dframe,ax=axis1)

sns.heatmap(flight_dframe,ax=axis2,cbar_kws={'orientation':'horizontal'})
```

```python
fig,(axis1,axis2) = plt.subplots(2,1)

yearly_flights=flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

years_dframe = pd.concat((years,flights),axis=1)
years_dframe.columns = ['Years','Flights']

sns.barplot('Years',y='Flights',data=years_dframe,ax=axis1)

sns.heatmap(flight_dframe,cmap='OrRd_r',ax=axis2,cbar_kws={'orientation':'horizontal'})
```
