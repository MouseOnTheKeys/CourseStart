```python
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = randn(100)

data2 = randn(100)
```

```python
sns.boxplot([data1,data2],color='orange')
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot([data2])
```

```python
data1 = stats.norm(0.5).rvs(100)

data2 = np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])

sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.1)
```

```python
sns.violinplot(data1,inner='stick')
```

```python
sns.violinplot(data2)
```
