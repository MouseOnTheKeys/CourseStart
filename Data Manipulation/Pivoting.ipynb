```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
#creating data import pandas testing utility
import pandas.util.testing as tm; tm.N = 3

#Create a unpivotal function 
def unpivot(frame):
    N,K = frame.shape
    
    data = {'value': frame.values.ravel('F'),'variable': np.asarray(frame.columns).repeat(N),'date': np.tile(np.asarray(frame.index),K)}
    # Return the DataFrame
    return DataFrame(data, columns=['date','variable','value'])

#set the DataFrame we`ll be useing
dframe = unpivot(tm.makeTimeDataFrame())
```
